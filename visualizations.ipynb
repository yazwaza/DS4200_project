```python
import pandas as pd

df = pd.read_csv('lung_cancer_prediction.csv')
df.head()
df.columns
```

```python
import altair as alt
```

Visualization 1 shows the age distribution as a density plot.

```python
df_sample = df.sample(n=5000, random_state=42)

alt.Chart(df_sample).transform_density(
    'Age',
    as_=['Age', 'density']
).mark_area(line=True, opacity=0.5).encode(
    x='Age:Q',
    y='density:Q'
).properties(
    width=500,
    height=300,
    title='Distribution of Age in the dataset'
)
```

The figure above shows that the density is similar and relatively flat for ages 40–80, roughly the age when lung function typically declines and lung disease risk rises. 

Visualization 2: Survival Probability by Different Stage Diagnosis

```python
stage_survival = df.groupby('Stage_at_Diagnosis', as_index=False)['5_Year_Survival_Probability'].mean()

alt.Chart(stage_survival).mark_bar().encode(
    x='Stage_at_Diagnosis:O',
    y='5_Year_Survival_Probability:Q',
    color='Stage_at_Diagnosis:O'
).properties(
    width=500,
    height=300,
    title='Survival Probability by Stage at Diagnosis'
)
```

Because this dataset was randomly generated without realistic correlations, survival probability does not vary meaningfully by stage and appears flat across diagnoses.
